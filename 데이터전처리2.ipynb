```python
import pandas as pd

# 기존 데이터 로드
df = pd.read_csv('processed_reviews.csv')

# 딥러닝 학습에 필요한 특징만 선택
cleaned_df = pd.DataFrame()

# 정규화된 텍스트 선택 (normalized_text가 더 정제되어 있어 학습에 적합)
cleaned_df['text'] = df['normalized_text']

# 유용한 수치형 특징 선택
cleaned_df['text_length'] = df['text_length']  # 텍스트 길이
cleaned_df['word_count'] = df['word_count']    # 단어 수
cleaned_df['special_char_count'] = df['special_char_count']  # 특수문자 수
cleaned_df['emoticon_count'] = df['emoticon_count']  # 이모티콘 수
cleaned_df['avg_words_per_sentence'] = df['avg_words_per_sentence']  # 문장당 평균 단어 수

# 감성 점수
cleaned_df['positive_score'] = df['positive_score']
cleaned_df['negative_score'] = df['negative_score']

# 저장
cleaned_df.to_csv('cleaned_reviews_for_learning.csv', index=False, encoding='utf-8-sig')
```
